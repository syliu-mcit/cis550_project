```python
lounge_missing = pd.DataFrame(columns = ['Percentage Missing'], data = lounge.apply(lambda x: (1-(x.count()/x.size))*100)).reset_index()
lounge_missing['Percentage Missing'] = round(lounge_missing['Percentage Missing'],2)
lounge_missing
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x="index", y="Percentage Missing", data = lounge_missing)
ax.bar_label(ax.containers[0])
plt.xticks(rotation=30, horizontalalignment="right")
plt.tight_layout()
```
