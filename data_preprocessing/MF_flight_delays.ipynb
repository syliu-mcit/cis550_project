```python
# ddl

CREATE TABLE FlightDelay AS (
    FL_DATE DATE,
    OP_CARRIER VARCHAR(30),
    OP_CARRIER_FL_NUM INT,
    ORIGIN VARCHAR(30),
    DEST VARCHAR(30),
    CRS_DEP_TIME INT,
    DEP_TIME INT,
    DEP_DELAY INT,
    TAXI_OUT INT,
    WHEELS_OFF INT,
    WHEELS_ON INT,
    TAXI_IN INT,
    CRS_ARR_TIME INT,
    ARR_TIME INT,
    ARR_DELAY INT,
    CANCELLED INT,
    CRS_ELAPSED_TIME INT,
    ACTUAL_ELAPSED_TIME INT,
    AIR_TIME INT,
    DISTANCE INT,
    TOTAL_DELAY INT,
    PRIMARY KEY (FL_DATE,OP_CARRIER,OP_CARRIER_FL_NUM)
)
```
