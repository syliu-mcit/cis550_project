## Airlines

```python
airlines_df = pd.read_csv('D:/550Project/proposal_flight/openflight/airlines.csv')
airlines_df.head()
```

```python
# number of nulls
airlines_df.isna().sum()
```

```python
# summary
airlines_df.describe()
```

```python
# only keep airline_id, name, iata, icao, country, active
# exclude airline_id < 0
airlines_res = airlines_df[['airline_id', 'name', 'iata', 'icao', 'country', 'active']]
airlines_res = airlines_res[airlines_res['airline_id'] > 0]
airlines_res.to_csv('D:/550Project/proposal_flight/openflight/airlines_res.csv', index=False)
airlines_res.head()
```

```python
# functional dependencies
airline_id -> name, iata, icao, country, active
```

```python
# ddl
CREATE TABLE Airlines AS(
    id int,
    name VARCHAR(30),
    iata VARCHAR(2),
    icao VARCHAR(3),
    country VARCHAR(50),
    active VARCHAR(5),
    PRIMARY KEY (id)
)
```

## Routes

```python
routes_df = pd.read_csv('D:/550Project/proposal_flight/openflight/routes.csv')
routes_df.head()
```

```python
# number of nulls
routes_df.isna().sum()
```

```python
# summary
routes_df.describe()
```

```python
# only keep airline, airline_id, source_airport, source_airport_id, destination_airport, destination_airport_id, stops, equipment
# exclude where airline_id is null
routes_res = routes_df[['airline', 'airline_id', 'source_airport', 'source_airport_id', 'destination_airport', 'destination_airport_id', 'stops', 'equipment']]
routes_res = routes_res[~routes_res['airline_id'].isna()]
routes_res.to_csv('D:/550Project/proposal_flight/openflight/routes_res.csv', index=False)
routes_res.head()
```

```python
# functional dependencies
airline_id -> airline
source_airport_id -> source_airport
destination_airport_id -> destination_airport
airline_id, source_airport_id, destination_airport_id -> stops, equipment

# can remove airline, source_airport, destination_airport and join with airlines, airports tables
```

```python
# ddl

CREATE TABLE Routes (
    airline VARCHAR(30),
    airlineID int,
    sourceAirport VARCHAR(30),
    sourceAirportID int,
    destinationAirport VARCHAR(30),
    destinationAirportID int,
    codeShare VARCHAR(30),
    stops int,
    equipment VARCHAR(30),
    PRIMARY KEY (airline, sourceAirportID, destinationAirportID))
    FOREIGN KEY (airlineID) references Airlines(id),
    (sourceAirportID) references Airports(id),
    (destinationAirportID) references Airport(id)
)
```
